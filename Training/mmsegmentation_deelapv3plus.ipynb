```python
#clean image directories to remove mask nad image where they do not coexist 
import os
for image_set in ['train', 'val']:
  images = os.listdir(os.path.join(img_dir, image_set))
  masks = os.listdir(os.path.join(ann_dir, image_set))
  for image in images:
    if image not in masks:
      os.remove(os.path.join(img_dir, image_set, image))
  for mask in masks:
    if mask not in images:
      os.remove(os.path.join(ann_dir, image_set, mask))
```
